```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer los datos desde un archivo CSV
datos = pd.read_csv('datos.csv', encoding='latin1', sep=';')

# Agrupar los datos por la variable "Residencia" y sumar los valores de "PartidosGanados"
datos_agrupados = datos.groupby('Residencia')['PartidosGanados'].sum()

# Crear un gr√°fico circular
plt.pie(datos_agrupados, labels=datos_agrupados.index)

# Agregar t√≠tulo al gr√°fico
plt.title('Distribuci√≥n de Partidos Ganados por Residencia')



# Mostrar el gr√°fico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer los datos desde un archivo CSV
datos = pd.read_csv('datos.csv', encoding='latin1', sep=';')

# Seleccionar las columnas de inter√©s
x = datos['Residencia']
y = datos['PartidosGanados']

# Crear un gr√°fico de dispersi√≥n
plt.scatter(x, y, s=5, alpha=1)

# Agregar etiquetas a los ejes
plt.xlabel('Residencia')
plt.ylabel('Partidos Ganados')

plt.scatter(x, y, s=5, alpha=1)

# Rotar etiquetas del eje x en 90 grados hacia la izquierda
plt.xticks(rotation=90)

plt.yticks(fontsize=5)
plt.tick_params(axis='y', which='both', pad=10)

# Mostrar el gr√°fico
plt.show()
```
