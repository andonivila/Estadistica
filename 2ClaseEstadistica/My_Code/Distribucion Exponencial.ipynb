```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return np.exp(x)
```

```python
x = np.linspace(-5, 5, 100)
```

```python
y = f(x)
```

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Función exponencial')
plt.grid(True)
plt.show()
```

```python
tasa = 0.5
valores = np.random.exponential(scale=1/tasa, size=10000)

# Crear un histograma de los valores
plt.hist(valores, bins=50, density=True, alpha=0.7, color='green')

# Etiquetas y título
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Distribución exponencial con tasa de 0.5')

# Mostrar el gráfico
plt.show()
```

```python
tasa = 0.5  # tasa de enfriamiento en grados por minuto
tasa_inversa = 1/tasa  # tasa inversa en minutos por grado

valores = np.random.exponential(scale=tasa_inversa, size=100000)
grados_faltantes = 50  # grados que faltan para enfriarse del todo
tiempo_total = 0
for valor in valores:
    tiempo_total += valor
    if tiempo_total >= grados_faltantes:
        break

    
print("El pastel tardará aproximadamente {:.2f} minutos en enfriarse de 70 grados a 20 grados".format(tiempo_total))
```

```python
# Crear un histograma de los valores
plt.hist(valores, bins=50, density=True, alpha=0.7, color='green')

# Calcular la curva de densidad de probabilidad teórica
x = np.linspace(0, max(valores), 1000)
pdf = tasa * np.exp(-tasa * x)
plt.plot(x, pdf, 'r', linewidth=2)

# Etiquetas y título
plt.xlabel('Tiempo (minutos)')
plt.ylabel('Densidad de probabilidad')
plt.title('Distribución exponencial con tasa de 0.5')

# Mostrar el gráfico
plt.show()
```
